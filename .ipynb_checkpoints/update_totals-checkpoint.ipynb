```python
import pandas as pd
import json
import os
```

```python
total_folder = 'Nuggets/!season_totals'
game_dir = 'Nuggets/10.27.23_Grizzlies'

for subdir, dirs, files in os.walk(game_dir):
    for file in files:
        if subdir == r'Nuggets/10.27.23_Grizzlies\Nuggets\Offense':
            overall_file = f'Nuggets/!season_totals/Nuggets/Offense/{file}'
            game_file = os.path.join(subdir, file)
            try:
               
                with open(game_file, 'r') as p:
                    game_data = json.load(p)

                with open(overall_file, 'r') as f:
                    overall_data = json.load(f)
                    
                new_game_data = game_data.copy()
                updated_overall_data = overall_data.copy()
                
                for key in overall_data.keys():
                    updated_overall_data[key] = pd.DataFrame(overall_data[key]['data']).transpose()
                for key in game_data.keys():
                    new_game_data[key] = pd.DataFrame(game_data[key]['data']).transpose()
                    
                for key in new_game_data.keys():
                    if key in updated_overall_data:
                        updated_overall_data[key] += new_game_data[key]
                    else:
                        updated_overall_data[key] = new_game_data[key]
                        
                for key in updated_overall_data.keys():
                    print(key, updated_overall_data[key])
                    updated_overall_data[key] = updated_overall_data[key].transpose().to_dict(orient='records')
                    
                print(updated_overall_data)
               # with open(f'NEW{file}', 'w', encoding='utf-8') as n:
                    #json.dump(updated_overall_data, n, ensure_ascii=False, indent=4)
                    
            except:
                print('No Overall Data')

```
